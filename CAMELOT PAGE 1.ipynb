```python
e = pd.DataFrame()
out = pd.DataFrame()
for i in range(1,240):
    file = str(i)+".pdf"
    print(file)
    table = camelot.read_pdf(file,pages='1',line_scale=60)
    l = pd.DataFrame()
    c = 0
    for i in table:
        df=table[c].df
        df = df.replace(to_replace ='', value = np.nan , regex = True) 
        df.dropna(axis=1, how='all',inplace=True)
        df.dropna(axis=0, how='all',inplace=True)
        df.dropna(axis=0, how='all',inplace=True)
        df.reset_index(inplace = True, drop = True)
        df.columns = range(df.shape[1])
        l=l.append(df)
        c=c+1
    df= l
    df.index = df.index + 1
    df_out = df.stack()
    df_out.index = df_out.index.map('{0[1]}_{0[0]}'.format)
    df= df_out.to_frame().T
    df = df.replace(to_replace ='[\n]', value = ' ', regex = True)
    df.columns = range(df.shape[1])
    a =df.shape[1]
    col =[]
    for i in range(0,a):
            if str(df[i])[5:11] in ['Delega','Title ','Contra','Specif','Vendor','Name a','City D','Term o','Maximu','Commit','Amendm']:
                pass
            else :
                col.append(i)
    df.drop(col,axis=1,inplace=True)
    df.columns = range(df.shape[1])
    a = df.shape[1]-10
    for i in range(0,a):
        df.drop([10+i],axis=1,inplace=True)
    df.columns = range(df.shape[1])
    df = df.replace(to_replace ='[\n]', value = ' ', regex = True) 
    if df.shape[1]<10:
        e = e.append(df)
    else :
        out =out.append(df)
out
```
